```python
for job in df['category'].unique():
    df[(df['category']==job ) &(df['gender']=='남') ].groupby(['age']).mean()[
        [str(i)+'|number' for i in range(1993,2020)]
    ].T.plot.line(figsize=(15,7), legend=True)
    plt.title(job)
    plt.show()
```

```python
df[(df['gender']=='남') ].groupby(['category']).mean()[
        [str(i)+'|number' for i in range(1993,2020)]
    ]
```

```python
for age in df['age'].unique():
    df[(df['gender']=='남') & (df['age']==age)].groupby(['category']).mean()[
        [str(i)+'|number' for i in range(1993,2020)]
    ].T.plot.line(figsize=(15,7), legend=True)
    plt.title(age)
    plt.show()
```

```python
for age in df['age'].unique():
    df[(df['gender']=='여') & (df['age']==age)].groupby(['category']).mean()[
        [str(i)+'|number' for i in range(1993,2020)]
    ].T.plot.line(figsize=(15,7), legend=True)
    plt.title(age)
    plt.show()
```
