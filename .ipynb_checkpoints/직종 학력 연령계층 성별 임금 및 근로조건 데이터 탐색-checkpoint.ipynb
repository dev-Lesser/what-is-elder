```python
from glob import glob
files = glob('./*.csv')
```

```python
df1 = pd.read_csv(files[0], encoding="CP949", header=[0,1])
df2 = pd.read_csv(files[1], encoding="CP949", header=[0,1])
df3 = pd.read_csv(files[2], encoding="CP949", header=[0,1])
```

```python
job_mapping ={}
for i in df1['직업별']['직업별'].unique():
    job_mapping[i] = ''
for i in df2['직업별']['직업별'].unique():
    job_mapping[i] = ''
for i in df3['한국표준직업분류6차']['한국표준직업분류6차'].unique():
    job_mapping[i] = ''
```

```python
for i in job_mapping:
    if i.find('관리자')>=0:
        job_mapping[i] = '관리자'
    elif i.find('전문가')>=0:
        job_mapping[i] = '전문가'
    elif i.find('사무')>=0:
        job_mapping[i] = '사무종사자'
    elif i.find('서비스')>=0:
        job_mapping[i] = '서비스-판매근로자'
    elif i.find('판매')>=0:
        job_mapping[i] = '서비스-판매근로자'
    elif i.find('농업')>=0 or i.find('농림')>=0:
        job_mapping[i] = '농업-어업숙련근로자'
    elif i.find('기능원')>=0:
        job_mapping[i] = '기능원'
    elif i.find('장치')>=0:
        job_mapping[i] = '장치기계조작-조립원'
    else :
        job_mapping[i] = '단순노무직'
        
```

```python
def mapping_jobs(d):
    d = d['직업별']
    return job_mapping[d]
df1['new_job_category'] =  df1['직업별'].apply(mapping_jobs, axis=1)
df2['new_job_category'] =  df2['직업별'].apply(mapping_jobs, axis=1)
```

```python
def mapping_jobs_recent(d):
    d = d['한국표준직업분류6차']
    return job_mapping[d]
df3['new_job_category'] =  df3['한국표준직업분류6차'].apply(mapping_jobs_recent, axis=1)
    
```

```python
from tqdm.notebook import tqdm
```

```python
import numpy as np
```

```python
def convert_nan(d):
    if d in [np.nan, '-']:
        return 0
    else: return d
```

```python
df1 = df1.applymap(convert_nan)
df2 = df2.applymap(convert_nan)
df3 = df3.applymap(convert_nan)
```

```python
df1[(df1['new_job_category']== '전문가')  ] [['new_job_category','성별','연령','1993','1994', '1995','1996','1997','1998','1999']]
```

```python
df = pd.DataFrame()
for gender in ['남','여']: # 남녀별 각각 통계
    for job in tqdm(list(set(job_mapping.values()))): # 각 직종별 통계
        tmp_df1 = df1[(df1['new_job_category']== job) & (df1['성별']['성별']== gender) ] [['new_job_category','성별','연령',
                                                                                      '1993','1994', '1995','1996','1997','1998','1999']]
        tmp_df1 = tmp_df1.reset_index(drop=True)
        new_columns = ['category','gender','age']
        for i in range(1993,2000):
            new_columns += [str(i)+'|mean_age',str(i)+'|working_time_all',str(i)+'|working_time_normal',str(i)+'|working_time_over',str(i)+'|number']
        tmp_df1.columns = new_columns
        if len(tmp_df1)==20:

            tmp1, tmp2 = tmp_df1.iloc[0:10], tmp_df1.iloc[10:]
            for i in tmp1.columns[3:]:
                tmp1 = tmp1.astype({i: float})
            for i in tmp2.columns[3:]:
                tmp2 = tmp2.astype({i: float})
            tmp1 = tmp1.reset_index(drop=True)
            tmp2 = tmp2.reset_index(drop=True)
#             print(job, tmp1)
            tmp_df1[new_columns[3:]] = tmp1[tmp1.columns[3:]].add(tmp2[tmp2.columns[3:]]) /2
        
        tmp_df1 = tmp_df1.dropna()
        tmp_df1 = tmp_df1.reset_index(drop=True)
        
        tmp_df2 = df2[(df2['new_job_category']== job) & (df2['성별']['성별']== gender) ][['new_job_category','성별','연령',
                                                                              '2000','2001','2002','2003','2004','2005','2006','2007','2008']]
        
        tmp_df2 = tmp_df2.reset_index(drop=True)
        new_columns = ['category','gender','age']
        for i in range(2000,2009):
            new_columns += [str(i)+'|mean_age',str(i)+'|working_time_all',str(i)+'|working_time_normal',str(i)+'|working_time_over',str(i)+'|number']
        tmp_df2.columns = new_columns
        if len(tmp_df2)==20:
            tmp1, tmp2 = tmp_df2.iloc[0:10], tmp_df2.iloc[10:]
            for i in tmp1.columns[3:]:
                tmp1 = tmp1.astype({i: float})
            for i in tmp2.columns[3:]:
                tmp2 = tmp2.astype({i: float})
            tmp1 = tmp1.reset_index(drop=True)
            tmp2 = tmp2.reset_index(drop=True)
            tmp_df2[new_columns[3:]] = tmp1[tmp1.columns[3:]].add(tmp2[tmp2.columns[3:]]) /2
        tmp_df2 = tmp_df2[new_columns[3:]]
        tmp_df2 = tmp_df2.dropna()
        tmp_df2 = tmp_df2.reset_index(drop=True)
        
        tmp_df3 = df3[(df3['new_job_category']== job) & (df3['성별']['성별']== gender) ][['new_job_category','성별','연령',
                                                                              '2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']]
        tmp_df3 = tmp_df3.reset_index(drop=True)
        new_columns = ['category','gender','age']
        for i in range(2009,2020):
            new_columns += [str(i)+'|mean_age',str(i)+'|working_time_all',str(i)+'|working_time_normal',str(i)+'|working_time_over',str(i)+'|number']
        tmp_df3.columns = new_columns
        if len(tmp_df3)==20:
            tmp1, tmp2 = tmp_df3.iloc[0:10], tmp_df3.iloc[10:]
            for i in tmp1.columns[3:]:
                tmp1 = tmp1.astype({i: float})
            for i in tmp2.columns[3:]:
                tmp2 = tmp2.astype({i: float})
            tmp1 = tmp1.reset_index(drop=True)
            tmp2 = tmp2.reset_index(drop=True)
            tmp_df3[new_columns[3:]] = tmp1[tmp1.columns[3:]].add(tmp2[tmp2.columns[3:]]) /2
        tmp_df3 = tmp_df3[new_columns[3:]]
        tmp_df3 = tmp_df3.dropna()
        tmp_df3 = tmp_df3.reset_index(drop=True)
        df = df.append(pd.concat([tmp_df1, tmp_df2, tmp_df3], axis=1))
        
```

```python
df.to_csv('./result/results.csv', encoding='CP949',index=False)
```
